```python
# create a function that divides the recipe num list into sets of 150 for API calls (daily limit for each key)
def divide_list(l, n): 
    
    # looping till length l 
    for i in range(0, len(l), n):  
        yield l[i:i + n] 

n = 1000  
x = list(divide_list(res, n)) 
recipe_list_1 = x[0]
recipe_list_2 = x[1]
recipe_list_3 = x[2]
recipe_list_4 = x[3]
recipe_list_5 = x[4]
recipe_list_6 = x[5]
recipe_list_7 = x[6]
recipe_list_8 = x[7]
recipe_list_9 = x[8]
recipe_list_10 = x[9]
```

```python
url_base = f"https://api.spoonacular.com/recipes/"
url_tail = "/information?includeNutrition=true"
key = f"&apiKey={api_key}"
# recipe_id = '100783'

s = requests.Session()
```

```python
data_1 = []

# Recipe Dataframe Components
rec_data_comps = ['id', 'title', 'readyInMinutes', 'nutrition>caloricBreakdown>percentProtein', 'nutrition>caloricBreakdown>percentFat','nutrition>caloricBreakdown>percentCarbs','nutrition>nutrients','servings','image','sourceUrl','summary',
                  'aggregateLikes','spoonacularScore','winePairing>pairedWines', 'winePairing>pairingText']

# create function to fetch the individual elements in the json data
def rec_comp_check(comp, recipe_data):
    try:
        comp_path = comp.split(">")
        for c in comp_path:
            recipe_data = recipe_data[c]
        datum = recipe_data
    except:
        datum = np.nan
    return datum
# run a loop to extract the recipe data
for num in recipe_list_1: # select the correct recipe list not exceeding api limit

    params = f'{num}'
    response = s.get(url_base + params + url_tail + key)
    if response.status_code == 200:
        recipe_data = response.json()
        my_dict ={"My ID": num}
        for comp in rec_data_comps:
            my_dict[comp] = rec_comp_check(comp, recipe_data)
        data_1.append(my_dict)
    else:
        print(f'{num} failed to return a good API response')
#         print(response)
```

```python
recipes_df_1 = data_1
df_one = pd.DataFrame(recipes_df_1)
df_one
```
