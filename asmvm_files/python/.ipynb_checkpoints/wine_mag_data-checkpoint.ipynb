```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv('csv/winemag-data-130k-v3.csv')
df
```

```python
df.info()
```

```python
us_wines = df.loc[(df['country'] == 'US')]
us_wines
```

```python
us_wines.columns
```

```python
usa_wines = us_wines.drop(['Unnamed: 0','designation','region_2', 'taster_name','taster_twitter_handle'], axis=1)
usa_wines
```

```python
america = usa_wines.loc[(usa_wines['province'] == 'America')]
america
```

```python
america['winery'].value_counts()
```

```python
filtered_df = usa_wines[usa_wines['province'] != 'America']
filtered_df = filtered_df[['country', 'province', 'winery', 'region_1', 'variety', 'title', 'description', 'points','price']]
# filtered_df['location'] = ''
filtered_df
```

```python
coords = pd.read_csv('csv/statelatlong.csv')
coords['location'] = ''
coords.head()
```

```python
# Use the lat/lng we recovered to identify airports
lat_lngs = []


for index, row in coords.iterrows():
    # get lat, lng from df
    lat = row["Latitude"]
    lng = row["Longitude"]
    lat_lngs = [lat, lng]
#     print(lat_lngs)
    try:
        coords.at[index, "location"] = lat_lngs
       
    except (KeyError, IndexError):
        print("this did not work for you.")
```

```python
coords.head()
```

```python
ca = coords[(coords['City'] == 'California')]
provinces = filtered_df['province']
provinces
```

```python
mapping = dict(coords[['City', 'location']].values)
filtered_df['location'] = filtered_df.province.map(mapping)
filtered_df
```

```python
filtered_df.reset_index(drop=True, inplace=True)
filtered_df
```

```python
filtered_df.to_csv('wine_ratings.csv')
```
